```python
import pandas as pd

df = pd.read_csv('kdrama_list_cleaned.csv')
```

```python
import re

def clean_img_url(url):
    if pd.isnull(url):
        return url
    # Keep only up to the image extension
    match = re.search(r'(.*\.(jpg|jpeg|png|webp))', url, re.IGNORECASE)
    if match:
        return match.group(1)
    else:
        return url
```

```python
df['img URL'] = df['img url'].apply(clean_img_url)
```
