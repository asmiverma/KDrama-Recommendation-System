# 02 - Exploratory Data Analysis
This notebook explores genre, ratings, tags, network, and yearly release trends.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from pathlib import Path

sns.set_theme(style='whitegrid')
df = pd.read_csv(Path('..') / 'data' / 'kdrama_list_cleaned.csv')
df = df.rename(columns={'img URL': 'img_URL'})
df.head()
```

```python
genres = (
    df['Genre'].fillna('')
    .str.split(',')
    .explode()
    .str.strip()
    .replace('', pd.NA)
    .dropna()
)
top_genres = genres.value_counts().head(15)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_genres.values, y=top_genres.index, palette='viridis')
plt.title('Top Genres')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Score'], bins=30, kde=True, color='teal')
plt.title('Rating Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

```python
tag_tokens = (
    df['Tags'].fillna('')
    .str.lower()
    .str.split(',')
    .explode()
    .str.strip()
    .replace('', pd.NA)
    .dropna()
)
top_tags = tag_tokens.value_counts().head(20)
top_tags
```

```python
if 'Network' in df.columns:
    network_counts = df['Network'].fillna('').replace('', pd.NA).dropna().value_counts().head(12)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=network_counts.values, y=network_counts.index, palette='magma')
    plt.title('Top Networks')
    plt.xlabel('Count')
    plt.ylabel('Network')
    plt.show()

year_counts = df['Year'].dropna().astype(int).value_counts().sort_index()
plt.figure(figsize=(10, 5))
sns.lineplot(x=year_counts.index, y=year_counts.values, marker='o')
plt.title('Drama Releases by Year')
plt.xlabel('Year')
plt.ylabel('Number of Dramas')
plt.show()
```
