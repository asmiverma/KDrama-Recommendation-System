# 01 - Data Cleaning
This notebook prepares the MyDramaList dataset for recommendation modeling.

```python
import pandas as pd
from pathlib import Path

DATA_PATH = Path('..') / 'data' / 'kdrama_list_cleaned.csv'
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
rename_map = {'img URL': 'img_URL', 'Img URL': 'img_URL', 'Image URL': 'img_URL'}
df = df.rename(columns=rename_map)

for col in ['Title', 'Synopsis', 'Genre', 'Tags', 'img_URL']:
    if col not in df.columns:
        df[col] = ''
    df[col] = df[col].fillna('').astype(str)

df['Year'] = pd.to_numeric(df.get('Year', 0), errors='coerce')
df['Score'] = pd.to_numeric(df.get('Score', 0), errors='coerce')
df = df.drop_duplicates(subset=['Title']).reset_index(drop=True)

df.shape
```

```python
df['combined_text'] = (
    df['Synopsis'].fillna('') + ' ' +
    df['Genre'].fillna('') + ' ' +
    df['Tags'].fillna('')
).str.strip()

output_path = Path('..') / 'data' / 'kdrama_list_cleaned_v2.csv'
df.to_csv(output_path, index=False)
output_path
```
